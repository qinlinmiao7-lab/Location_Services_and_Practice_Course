如何在linux远程服务器上下载ROS (tested with Noetic)？

注意：ROS一定要在根目录路径下下载

一、先确认你的服务器环境
ROS (tested with Noetic) 指的是 ROS Noetic 版本，它只支持 Ubuntu 20.04 (Focal) 系统。
你可以先在 SSH 里执行下面命令确认系统版本：

```python
# shellcheck disable=SC2148
lsb_release -a
```

如果输出里 Release 不是 20.04，就不能直接安装 Noetic，需要考虑其他方案（比如 Docker）。

二、在 Ubuntu 20.04 上安装 ROS Noetic
在 SSH 里依次执行以下命令：
1. 设置软件源

```python
sudo sh -c 'echo "deb http://packages.ros.org/ros/ubuntu $(lsb_release -sc) main" > /etc/apt/sources.list.d/ros-latest.list'
```

2. 添加密钥

```python
sudo apt install curl -y
curl -s https://raw.githubusercontent.com/ros/rosdistro/master/ros.asc | sudo apt-key add -
```

3. 更新包列表

```python
sudo apt update
```

4. 安装 ROS Noetic

```python
sudo apt install ros-noetic-desktop-full -y
```

5. 配置环境变量

```python
echo "source /opt/ros/noetic/setup.bash" >> ~/.bashrc
source ~/.bashrc
```

6. 安装依赖工具

```python
sudo apt install python3-rosdep python3-rosinstall python3-rosinstall-generator python3-wstool build-essential -y
sudo rosdep init
rosdep update
```

如果是下载基于 ROS 的项目代码并编译，必须在 ROS 工作空间（workspace）目录下操作，步骤如下：

```python
# 创建工作空间目录（名字一般叫 catkin_ws，也可以自定义）
mkdir -p catkin_ws/src
# 进入工作空间
cd catkin_ws/src
```

mkdir -p catkin_ws/src：-p 表示自动创建多级目录，catkin_ws 是工作空间根目录，src 是存放源码的子目录（必须叫 src）；
工作空间建议放在 /home/你的用户名/ 下（比如 /home/ubuntu/catkin_ws），避免放在系统目录（如 /root、/opt）导致权限问题。
